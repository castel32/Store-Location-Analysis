### 2. Choose Relevant Columns

#### 2.1. Zip

```python
#When previewing the data, the first lines had Zipcode==0. They are summary lines for each of the states.
print('zipcodes == 0:',df[df['ZIPCODE']==0].shape) #output: (306, 167)
l=df[df['ZIPCODE']==0]['STATE'].unique().tolist()
print(l)
print(len(l), "unique States")
#(306/6lines per zip) #output: 51 States. There is one summary for each state. Zipcode==0.
# Note PUERTO RICO is not in this IRS list. Keep in mind when we join Zip codes.

#Removing those summary state lines.
df=df[df['ZIPCODE']!=0]
print("Removed State summary lines")

#Importing the first column into the new DF.
irs11=pd.DataFrame()
irs11['zip']=df['ZIPCODE'].astype(str)
#show me the lenghts of zip code
irs11['zip'].str.len().unique() #output: 4 and 5, we need to fix the lenght 4. They are missing a 0.
#Adding a leading '0'.
irs11.loc[irs11['zip'].str.len() == 4,'zip'] = '0' + irs11['zip']
print('4 digit Zips adjusted to 5 digit')
```

#### 2.2. State

```python
irs11['state']=df['STATE']
irs11['state'].describe()
```

#### 2.3. AGI_Stub 
The AGI_stub categorization is used to group tax returns or taxpayers into different income ranges for statistical analysis.

```python
irs11['agi_stub']=df['agi_stub']
irs11['agi_stub'].unique()
#irs11['agi_stub'].describe()
```

#### 2.3. Tax Fields

##### 2.3.1. Number of returns

```python
irs11['n_returns']=df['N1'].astype(int)
irs11['n_returns'].describe()
#There are some 0 present. Let's investigate it
total_zips=irs11['zip'].nunique()
total_zips_with_zero_returns=irs11[irs11['n_returns']==0]['zip'].nunique()

print('{:.2f}'.format(total_zips_with_zero_returns/total_zips*100)  ,'% zip codes have at least one agi_stub with 0 returns')
```

#### 2.4. Data transformations 

##### 2.4.1. Pivot 'Number of Returns' per agi_stub

```python
# Pivot the data to have AGI_stub as columns and ZIP codes as rows, with returns as values
pivot_n_returns = irs11.pivot_table(index='zip', columns='agi_stub', values='n_returns', fill_value=0)

# Calculate the total number of returns for each ZIP code
pivot_n_returns['total_returns'] = pivot_n_returns.sum(axis=1)

# Calculate the ratio of n_returns for each AGI_stub within each ZIP code
for column in pivot_n_returns.columns[:-1]:  # Exclude the 'total_returns' column
    pivot_n_returns[column] = pivot_n_returns[column] / pivot_n_returns['total_returns']

current_columns = pivot_n_returns.columns.tolist()
print(current_columns)
new_columns = ['total_returns'] + [col for col in current_columns if col != 'total_returns']
pivot_n_returns = pivot_n_returns[new_columns]

# Remove the name of the columns index
pivot_n_returns.columns.name = None
# Reset the index so zip is a column again.
pivot_n_returns.reset_index(inplace=True)
#print(pivot_n_returns.columns)
pivot_n_returns.rename(columns={1 : 'agi_stub_1_ratio',2 : 'agi_stub_2_ratio',3 : 'agi_stub_3_ratio',4 : 'agi_stub_4_ratio',5 : 'agi_stub_5_ratio',6 : 'agi_stub_6_ratio',},inplace=True)
pivot_n_returns
```

##### 2.4.3. Merge the data together

```python
# First merge the n_returns ratios
newdf1=pd.merge(irs11[['zip','state']],pivot_n_returns, how='right', on='zip')
newdf1.drop_duplicates(inplace=True)

#Let's update irs11 and update it
irs11=newdf1
del newdf1

irs11.head()
```
