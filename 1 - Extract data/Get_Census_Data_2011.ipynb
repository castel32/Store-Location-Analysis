### 2. Choose Relevant Columns

#### 2.1. Zip

```python
#Census 2011 variable is c11 for quickness
c11=pd.DataFrame()

#Zip
c11['zip']=df['NAME'].str.replace('ZCTA5 ','')
#Check Lenght of Zip codes is correct
c11[(c11['zip'].str.len() != 5)]
```

#### 2.2. Sex

```python
#Sex -------
c11['total_pop']=df['DP05_0001E'].astype(int)
c11['male_pop']=df['DP05_0002E'].astype(int)
c11['female_pop']=df['DP05_0003E'].astype(int)
#Sex Ratios
c11['male_pop_ratio']=c11['male_pop']/c11['total_pop']
c11['female_pop_ratio']=c11['female_pop']/c11['total_pop']
```

##### 2.2.1. Verification

```python
#Test Sex (I want to see if the sum of the ratios ==1)
c11['sum_sex_ratios']=c11['male_pop_ratio']+c11['female_pop_ratio']
c11['diff_total_sex']=1-c11['sum_sex_ratios']
print(c11['diff_total_sex'].sum())
if(c11['diff_total_sex'].sum() == 0):
    c11.drop(columns=['diff_total_sex'], inplace=True)
    c11.drop(columns=['sum_sex_ratios'], inplace=True)
    
```

#### 2.3. Age Groups

```python
#Ages -------
c11['age_under5_pop']=df['DP05_0004E'].astype(int)
c11['age_5to9_pop']=df['DP05_0005E'].astype(int)
c11['age_10to14_pop']=df['DP05_0006E'].astype(int)
c11['age_15to19_pop']=df['DP05_0007E'].astype(int)
c11['age_20to24_pop']=df['DP05_0008E'].astype(int)
c11['age_25to34_pop']=df['DP05_0009E'].astype(int)
c11['age_35to44_pop']=df['DP05_0010E'].astype(int)
c11['age_45to54_pop']=df['DP05_0011E'].astype(int)
c11['age_55to59_pop']=df['DP05_0012E'].astype(int)
c11['age_60to64_pop']=df['DP05_0013E'].astype(int)
c11['age_65to74_pop']=df['DP05_0014E'].astype(int)
c11['age_75to84_pop']=df['DP05_0015E'].astype(int)
c11['age_85andOver_pop']=df['DP05_0016E'].astype(int)
#Ages Ratios
c11['age_under5_pop_ratio'] = c11['age_under5_pop']/c11['total_pop']   
c11['age_5to9_pop_ratio'] = c11['age_5to9_pop']/c11['total_pop']      
c11['age_10to14_pop_ratio'] = c11['age_10to14_pop']/c11['total_pop']
c11['age_15to19_pop_ratio'] = c11['age_15to19_pop']/c11['total_pop']
c11['age_20to24_pop_ratio'] = c11['age_20to24_pop']/c11['total_pop']
c11['age_25to34_pop_ratio'] = c11['age_25to34_pop']/c11['total_pop']
c11['age_35to44_pop_ratio'] = c11['age_35to44_pop']/c11['total_pop']
c11['age_45to54_pop_ratio'] = c11['age_45to54_pop']/c11['total_pop']
c11['age_55to59_pop_ratio'] = c11['age_55to59_pop']/c11['total_pop']
c11['age_60to64_pop_ratio'] = c11['age_60to64_pop']/c11['total_pop']
c11['age_65to74_pop_ratio'] = c11['age_65to74_pop']/c11['total_pop']
c11['age_75to84_pop_ratio'] = c11['age_75to84_pop']/c11['total_pop']
c11['age_85andOver_pop_ratio'] = c11['age_85andOver_pop']/c11['total_pop']
```

##### 2.3.1 Verification

```python
c11['sum_age_ratios']=(c11['age_under5_pop_ratio']+c11['age_5to9_pop_ratio']+c11['age_10to14_pop_ratio']+c11['age_15to19_pop_ratio']+c11['age_20to24_pop_ratio']+c11['age_25to34_pop_ratio']+c11['age_35to44_pop_ratio']+c11['age_45to54_pop_ratio']+c11['age_55to59_pop_ratio']+c11['age_60to64_pop_ratio']+c11['age_65to74_pop_ratio']+c11['age_75to84_pop_ratio']+c11['age_85andOver_pop_ratio'])
```

##### 2.3.2 Verification  - Potential incomplete age groups data.

```python
# Due to missing an age interval many zip codes failed to sum 1 when adding all the ages ratios.

# This is the work that was performed to analyze at that moment.

# How much detail are we missing in those zipcodes that the sum of age ratios is less than 1? 

import pandas as pd
import matplotlib.pyplot as plt

# Assuming c11 is your DataFrame

# Create the boxplot
plt.figure(figsize=(8, 6))
boxplot = c11['sum_age_ratios'].plot(kind='box')

# Calculate mean and quartiles
mean = c11['sum_age_ratios'].mean()
q1 = c11['sum_age_ratios'].quantile(0.25)
q3 = c11['sum_age_ratios'].quantile(0.75)
median = c11['sum_age_ratios'].median()

# Add mean and quartiles to the plot
plt.text(0.95, mean, f'Mean: {mean:.2f}', fontsize=10, va='center', ha='left')
plt.text(0.95, q1, f'Q1: {q1:.2f}', fontsize=10, va='center', ha='left')
plt.text(0.95, median, f'Median: {median:.2f}', fontsize=10, va='center', ha='left')
plt.text(0.95, q3, f'Q3: {q3:.2f}', fontsize=10, va='center', ha='left')

plt.title("Boxplot of 'sum_age_ratios' with Mean and Quartiles")
plt.xlabel("sum_age_ratios")
plt.ylabel("Values")

# Show the plot
plt.show()
```

##### 2.3.3 Verification  - Quantify the outliers

```python
Q1 = c11['sum_age_ratios'].quantile(0.25)
Q3 = c11['sum_age_ratios'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print('Lower Outlier limit',lower_bound)
print('Upper Outlier limit',upper_bound)

print('Very small STD: ',c11['sum_age_ratios'].std())

# Count the number of outliers
outliers_count = ((c11['sum_age_ratios'] < lower_bound) | (c11['sum_age_ratios'] > upper_bound)).sum()

print("Number of outliers:", outliers_count/c11.shape[0]*100,'%')

#Let's filter the outliers and mark them:
filter=(c11['sum_age_ratios'] < lower_bound) | (c11['sum_age_ratios'] > upper_bound)
c11['age_group_outlier']=(c11['sum_age_ratios'] < lower_bound) | (c11['sum_age_ratios'] > upper_bound)

#The number of outliers is high but the STD is so small that it is not a concern.
# Moving forward we don't need the column: 
c11.drop(columns=['sum_age_ratios'], inplace=True)
```

#### 2.4. Race

```python
#Race
c11['race_total_pop']=df['DP05_0028E'].astype(int)
#One or Two or More races
c11['race_one']=df['DP05_0029E'].astype(int)
c11['race_two_or_more']=df['DP05_0030E'].astype(int)

#Breakdown of Ethnicity & Race
c11['race_hispanic_or_latino']=df['DP05_0066E'].astype(int)
c11['race_white_alone']=df['DP05_0072E'].astype(int)
c11['race_black_or_african']=df['DP05_0073E'].astype(int)
c11['race_american_indian_and_alaska_native']=df['DP05_0074E'].astype(int)
c11['race_asian']=df['DP05_0075E'].astype(int)
c11['race_native_hawaian_and_other_pacific_islander']=df['DP05_0076E'].astype(int)
c11['race_some_other_race_alone']=df['DP05_0077E'].astype(int)
c11['race_two_or_more_races']=df['DP05_0078E'].astype(int)

```

##### 2.4.1. Verifications

```python
#Does this add any info?
#Checking if 'total_pop','race_total_pop' have the same info.
c11['total_race_diff']=c11['total_pop']-c11['race_total_pop']
#c11[['total_pop','race_total_pop']]
if(c11[c11['total_race_diff']!=0].empty == True):
    print('''Check 1: ['total_pop'] and ['race_total_pop'] contain the same information''')
    #Result: They do, they are the same. I will not include this column.
    c11.drop(columns=['race_total_pop'], inplace=True)
    c11.drop(columns=['total_race_diff'], inplace=True)


#Does the breakdown for ethnicity/race add up to the total_pop?
c11['sum_races']=(c11['race_hispanic_or_latino']+c11['race_white_alone']+c11['race_black_or_african']+c11['race_american_indian_and_alaska_native']+c11['race_asian']+c11['race_native_hawaian_and_other_pacific_islander']+c11['race_some_other_race_alone']+c11['race_two_or_more_races'])
c11['diff_sum_races']=c11['total_pop']-c11['sum_races']
if(c11[c11['diff_sum_races']!=0].empty == True):
    print('''Check 2: ['total_pop'] and ['sum_races'] contain the same information''')
    #Result: The breakdown is perfect for all zip codes. We remove the test columns
    c11.drop(columns=['sum_races'], inplace=True)
    c11.drop(columns=['diff_sum_races'], inplace=True)

```
